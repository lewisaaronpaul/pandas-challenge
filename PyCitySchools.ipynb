```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data = school_data.sort_values(by = 'school_name').reset_index(drop=True)
school_data.head()
```

```python
student_data = student_data.sort_values(by = 'school_name').reset_index(drop=True)
student_data.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
district_summary = dict()
```

```python
#Calculate the total number of schools
district_summary['Total Schools'] = len(school_data['school_name'])
```

```python
#Calculate the total number of students
total_students = school_data['size'].sum()
district_summary['Total Students'] = format(total_students, ",d")
```

```python
#Calculate the total budget
total_budget = school_data['budget'].sum()
district_summary['Total Budget'] = "$" + format(total_budget, ",.2f")
```

```python
#Calculate the average math score
avg_math_score = student_data['math_score'].mean()
district_summary['Average Math Score'] = format(avg_math_score, ".2f")
```

```python
#Calculate the average reading score
avg_reading_score = student_data['reading_score'].mean()
district_summary['Average Reading Score'] = format(avg_reading_score, ".2f")
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = student_data['math_score'] >= 70
passing_math_df = student_data[passing_math]
total_passing_math = len(passing_math_df)
district_summary['% Passing Math'] = format(total_passing_math * 100 / total_students, ".2f")
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading = student_data['reading_score'] >= 70
passing_reading_df = student_data[passing_reading]
total_passing_reading = len(passing_reading_df)
district_summary['% Passing Reading'] = format(total_passing_reading * 100 / total_students, ".2f")
```

```python
#Calculate the percentage of students who passed math **and** reading (% Overall Passing)
passing_math_reading = passing_math & passing_reading
passing_math_reading_df = student_data[passing_math_reading]
total_passing_math_reading = len(passing_math_reading_df)
district_summary['% Overall Passing'] = format(total_passing_math_reading * 100 / total_students, ".2f")
```

```python
#Create a dataframe to hold the above results
district_summary_df = pd.DataFrame(district_summary, index = [0])
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
grp_name_type = school_data_complete.groupby(['school_name', 'type'])
```

```python
school_summary_df = grp_name_type[["student_name"]].count().reset_index()
school_summary_df.rename(columns = {'school_name' : 'school_name', 'type' : 'School Type', 'student_name' : 'Total Students' }, inplace = True) 
school_summary_df.set_index('school_name', inplace = True)
school_summary_df.head()
```

```python
school_data.columns
```

```python
#Total School Budget
school_summary_df['Total School Budget'] = school_data['budget'].values
```

```python
#Per Student Budget
school_summary_df['Per Student Budget'] = school_summary_df['Total School Budget'] / school_summary_df['Total Students']
school_summary_df.head(2)
```

```python
#Average Math Score
school_summary_df['Average Math Score'] = student_data.groupby('school_name').math_score.mean()
```

```python
#Average Reading Score
school_summary_df['Average Reading Score'] = student_data.groupby('school_name').reading_score.mean()
```

```python
#The percentage of students that passed math.
#Create a new DataFrame for students passing math. 
passing_math_df = student_data[student_data["math_score"] >= 70]
school_summary_df['% Passing Math'] = passing_math_df.groupby('school_name').math_score.count() * 100 / school_summary_df['Total Students']
```

```python
#The percentage of students that passed reading.
#Create a new DataFrame for students passing reading. 
passing_reading_df = student_data[student_data["reading_score"] >= 70]
school_summary_df['% Passing Reading'] = passing_reading_df.groupby('school_name').reading_score.count() * 100 / school_summary_df['Total Students']
```

```python
#% Overall Passing (The percentage of students that passed math and reading.)
#Create a new DataFrame for students passing math and reading. 
passing_math_reading_df = student_data[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70)]
school_summary_df['% Overall Passing'] = passing_math_reading_df.groupby('school_name').reading_score.count() * 100 / school_summary_df['Total Students']
```

```python
# Format the decimal places, include a dollar sign, and use comma notation
school_summary_df_copy = school_summary_df.copy()
school_summary_df_copy["Total School Budget"] = school_summary_df_copy["Total School Budget"].map("${:,.2f}".format)
school_summary_df_copy["Per Student Budget"] = school_summary_df_copy["Per Student Budget"].map("${:,.2f}".format)

school_summary_df_copy["Average Math Score"] = school_summary_df_copy["Average Math Score"].map("{:.2f}".format)
school_summary_df_copy["Average Reading Score"] = school_summary_df_copy["Average Reading Score"].map("{:.2f}".format)
school_summary_df_copy["% Passing Math"] = school_summary_df_copy["% Passing Math"].map("{:.2f}".format)
school_summary_df_copy["% Passing Reading"] = school_summary_df_copy["% Passing Reading"].map("{:.2f}".format)
school_summary_df_copy["% Overall Passing"] = school_summary_df_copy["% Overall Passing"].map("{:.2f}".format)

school_summary_df_copy.index.set_names(names = '', inplace = True)
school_summary_df_copy
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
#Use the pivot_table function to list the average Math Score for students of each 
#grade level (9th, 10th, 11th, 12th) at each school.

df_math = school_data_complete.pivot_table(values = 'math_score',
                                index = 'school_name',
                                columns = 'grade',
                                aggfunc = 'mean')
columns = ['9th', '10th', '11th', '12th']
df_math_reindex = df_math.reindex(columns = columns)

df_math_reindex["9th"] = df_math_reindex["9th"].map("{:.2f}".format)
df_math_reindex["10th"] = df_math_reindex["10th"].map("{:.2f}".format)
df_math_reindex["11th"] = df_math_reindex["11th"].map("{:.2f}".format)
df_math_reindex["12th"] = df_math_reindex["12th"].map("{:.2f}".format)

df_math_reindex.index.set_names(names = '', inplace = True)
df_math_reindex
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create a copy of school_data_complete DataFrame
bins_df = school_data_complete.copy()
```

```python
# Calculate spending per student.
bins_df['Spending per Std'] = bins_df['budget'] / bins_df['size']
```

```python
bins = [0, 584, 629, 644,  675]
group_names = ["<$584", "$585-629","$630-644","$645-675"]
```

```python
#DataFrame with 'Spending per Std' column and 'Spending Ranges (Per Student)' column.
cuts = pd.cut(bins_df["Spending per Std"], bins, labels = group_names, include_lowest = True)
bins_df["Spending Ranges (Per Student)"] = cuts
bins_df.head()
```

```python
# Average Math Score and Average Reading Score.
school_spending_df = bins_df.groupby(['Spending Ranges (Per Student)']).agg({'math_score' : 'mean',
                                                                           'reading_score' : 'mean'})
school_spending_df.rename(columns = {'Spending Ranges (Per Student)' : 'Spending Ranges (Per Student)',
                                    'math_score' : 'Average Math Score',
                                    'reading_score' : 'Average Reading Score' }, inplace = True) 
```

```python
# Number of students in each spending range.
spending_range_count = bins_df.groupby('Spending Ranges (Per Student)').student_name.count()
spending_range_count
```

```python
#The percentage of students that passed math.
#Create a new DataFrame for students passing math. 
passing_math = bins_df[bins_df["math_score"] >= 70]

school_spending_df['% Passing Math'] = passing_math.groupby('Spending Ranges (Per Student)').math_score.count() * 100 / spending_range_count
```

```python
#The percentage of students that passed reading.
#Create a new DataFrame for students passing reading. 
passing_reading = bins_df[bins_df["reading_score"] >= 70]

school_spending_df['% Passing Reading'] = passing_reading.groupby('Spending Ranges (Per Student)').reading_score.count() * 100 / spending_range_count
```

```python
#% Overall Passing (The percentage of students that passed math and reading.)
#Create a new DataFrame for students passing math and reading. 
passing_math_reading = bins_df[(bins_df["math_score"] >= 70) & (bins_df["reading_score"] >= 70)]
school_spending_df['% Overall Passing'] = passing_math_reading.groupby('Spending Ranges (Per Student)').reading_score.count() * 100 / spending_range_count
```

```python
# Format the decimal places, include a dollar sign, and use comma notation
school_spending_df_copy = school_spending_df.copy()
school_spending_df_copy["Average Math Score"] = school_spending_df_copy["Average Math Score"].map("{:.2f}".format)
school_spending_df_copy["Average Reading Score"] = school_spending_df_copy["Average Reading Score"].map("{:.2f}".format)
school_spending_df_copy["% Passing Math"] = school_spending_df_copy["% Passing Math"].map("{:.2f}".format)
school_spending_df_copy["% Passing Reading"] = school_spending_df_copy["% Passing Reading"].map("{:.2f}".format)
school_spending_df_copy["% Overall Passing"] = school_spending_df_copy["% Overall Passing"].map("{:.2f}".format)

school_spending_df_copy
```
